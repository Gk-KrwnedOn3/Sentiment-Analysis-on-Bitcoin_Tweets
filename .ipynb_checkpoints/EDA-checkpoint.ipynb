```python
import pandas as pd
import emoji
import re

# Load the dataset (replace 'tweets.csv' with the actual path to your file)
df = pd.read_csv('tweets.csv')

# Function to remove URLs
def remove_urls(text):
    url_pattern = r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
    return re.sub(url_pattern, "", text)

# Function to remove symbols
def remove_symbols(text):
    return re.sub(r"[^\w\s]", "", text)  # Keeps only words and spaces

# Function to remove emojis
def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"  # dingbats
        u"\U000024C2-\U0001F251"
        u"\U00010000-\U0010ffff"
        u"\u200d"                 # Zero-width joiner
        u"\u2640-\u2642"          # gender symbols
        u"\u2600-\u2B55"          # miscellaneous symbols
        u"\u23cf"                 # eject symbol
        u"\u23e9"                 # fast-forward and rewind symbols
        u"\u231a"                 # watch
        u"\ufe0f"                 # dingbats
        u"\u3030"
        "]+", flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Clean the text column
df['cleaned_text'] = df['text'].apply(remove_urls).apply(remove_symbols).apply(remove_emojis)

# Check the cleaned data
print(df[['text', 'cleaned_text']].head())
```

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd

# Download historical Bitcoin price data from Yahoo Finance
btc_data = yf.download("BTC-USD", start="2022-01-01", end="2023-06-22", interval='1d')

# Reset index to ensure 'Date' is a column
btc_data.reset_index(inplace=True)

# Keep only 'Date' and 'Close' columns (closing price)
btc_data = btc_data.loc[:, ['Date', 'Close']].copy()

# Rename the 'Close' column to 'Bitcoin_Price' for clarity
btc_data.rename(columns={'Close': 'Bitcoin_Price'}, inplace=True)

# Convert the 'Date' column to datetime format
btc_data['Date'] = pd.to_datetime(btc_data['Date'])

# Format the 'Date' column to 'DD/MM/YYYY'
btc_data['Date'] = btc_data['Date'].dt.strftime('%d/%m/%Y')

# Save the Bitcoin price data to a CSV file
btc_data.to_csv('bitcoin_price_2022_2023.csv', index=False)

# Print the first few rows of the Bitcoin price data to verify
print(btc_data.head())
```

```python
# Load the TextBlob and VADER sentiment datasets
textblob_df = pd.read_csv('tweets_with_textblob_sentiment.csv')
vader_df = pd.read_csv('tweets_with_vader_sentiment.csv')

# Ensure the 'date' column is in datetime format for proper plotting, with a specified format
textblob_df['date'] = pd.to_datetime(textblob_df['date'], format='%d/%m/%Y')
vader_df['date'] = pd.to_datetime(vader_df['date'], format='%d/%m/%Y')

# Aggregate sentiment scores by date (average sentiment per day)
textblob_daily = textblob_df.groupby('date')['sentiment_score'].mean().reset_index()
vader_daily = vader_df.groupby('date')['sentiment_score'].mean().reset_index()

# Ensure the 'Date' column in the Bitcoin dataset is in datetime format
btc_data['Date'] = pd.to_datetime(btc_data['Date'], format='%d/%m/%Y')

# Merge TextBlob, VADER, and Bitcoin data on 'date'
merged_data = pd.merge(textblob_daily, vader_daily, on='date', how='inner', suffixes=('_textblob', '_vader'))
merged_data = pd.merge(merged_data, btc_data, left_on='date', right_on='Date', how='inner')

# Drop the extra 'Date' column
merged_data.drop('Date', axis=1, inplace=True)

# Convert 'date' column to 'DD/MM/YYYY' format
merged_data['date'] = merged_data['date'].dt.strftime('%d/%m/%Y')

# Show the first few rows of merged data
print(merged_data.head())
```

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Bitcoin price on the first y-axis (left side)
ax1.set_xlabel('Date')
ax1.set_ylabel('Bitcoin Price (USD)', color='red')
ax1.plot(merged_data['date'], merged_data['Bitcoin_Price'], label='Bitcoin Price', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis (right side) that shares the same x-axis for sentiment
ax2 = ax1.twinx()
ax2.set_ylabel('Sentiment Score', color='blue')

# Plot TextBlob sentiment
ax2.plot(merged_data['date'], merged_data['sentiment_score_textblob'], label='TextBlob Sentiment', color='blue')

# Plot VADER sentiment
ax2.plot(merged_data['date'], merged_data['sentiment_score_vader'], label='VADER Sentiment', color='green')

# Customize the second y-axis
ax2.tick_params(axis='y', labelcolor='blue')

# Add titles and rotate x-axis labels for readability
plt.title('Bitcoin Price vs Sentiment (TextBlob & VADER)')
plt.xticks(rotation=45)

# Add a legend to the figure
fig.tight_layout()

# Show the plot
plt.show()
```

```python
# Normalize the Bitcoin price and sentiment scores using Min-Max scaling
merged_data['Bitcoin_Price_normalized'] = (merged_data['Bitcoin_Price'] - merged_data['Bitcoin_Price'].min()) / (merged_data['Bitcoin_Price'].max() - merged_data['Bitcoin_Price'].min())
merged_data['sentiment_score_textblob_normalized'] = (merged_data['sentiment_score_textblob'] - merged_data['sentiment_score_textblob'].min()) / (merged_data['sentiment_score_textblob'].max() - merged_data['sentiment_score_textblob'].min())
merged_data['sentiment_score_vader_normalized'] = (merged_data['sentiment_score_vader'] - merged_data['sentiment_score_vader'].min()) / (merged_data['sentiment_score_vader'].max() - merged_data['sentiment_score_vader'].min())

# Plot normalized data
plt.figure(figsize=(10, 6))

# Plot normalized TextBlob sentiment
plt.plot(merged_data['date'], merged_data['sentiment_score_textblob_normalized'], label='TextBlob Sentiment (Normalized)', color='blue')

# Plot normalized VADER sentiment
plt.plot(merged_data['date'], merged_data['sentiment_score_vader_normalized'], label='VADER Sentiment (Normalized)', color='green')

# Plot normalized Bitcoin price
plt.plot(merged_data['date'], merged_data['Bitcoin_Price_normalized'], label='Bitcoin Price (Normalized)', color='red')

# Add titles and labels
plt.title('Bitcoin Price vs Sentiment (Normalized)')
plt.xlabel('Date')
plt.ylabel('Normalized Sentiment/Price (0-1)')
plt.legend()

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

1. Normalization (Min-Max Scaling):
Best suited for: Visualizing trends and comparing data between Bitcoin prices and sentiment over time.
Why: It scales both sentiment scores and Bitcoin prices to a common range (e.g., 0-1), which allows you to easily compare the relative movements of sentiment and Bitcoin prices, even though they are on different scales.
Use case: When you want to see how Bitcoin price movements and sentiment scores change in relation to each other

```python
# Calculate correlation between Bitcoin prices and sentiment scores
correlation_textblob = merged_data[['Bitcoin_Price', 'sentiment_score_textblob']].corr().iloc[0, 1]
correlation_vader = merged_data[['Bitcoin_Price', 'sentiment_score_vader']].corr().iloc[0, 1]

print(f"Correlation between Bitcoin Price and TextBlob Sentiment: {correlation_textblob}")
print(f"Correlation between Bitcoin Price and VADER Sentiment: {correlation_vader}")
```

2. Correlation Analysis:
Best suited for: Measuring the relationship between sentiment scores and Bitcoin prices.
Why: Correlation quantifies the degree of association between sentiment and price, ranging from -1 (strong negative correlation) to +1 (strong positive correlation). This helps determine whether an increase in sentiment (positive or negative) is associated with a rise or fall in Bitcoin prices.
Use case: If you want to assess whether Bitcoin prices are directly influenced by public sentiment or vice versa.
